```python
class Employees:
    def __init__(self, first, last, pay):
        self.first=first
        self.last=last
        self.pay=pay
        self.mail=first+'.'+last+'@company.com'
emp_1=Employees('severthy','sneha',50000)
emp_2=Employees('pooja','reddy',60000)
print(emp_1.mail)
print(emp_2.mail)
```

```python
class Employ:
    def __init__(self, first, last, pay):
        self.first=first
        self.last=last
        self.pay=pay
        self.mail=first+'.'+last+'@company.com'
    def fullname(self):
        return '{}{}'.format(self.first,self.last)
emp_1=Employ('severthy','sneha',50000)
emp_2=Employ('pooja','reddy',60000)
print(emp_1.mail)
print(emp_2.mail)
print(emp_1.fullname())
```

```python
class Student:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def show(self):
        print(f"Name:{self.name}")
        print(f"Age:{self.age}")
s1=Student("RM",29)
s1.show()
```

```python
class Employee:
    def __init__(self,name,salary):
        self.name=name
        self.salary=salary
    def increase_salary(self):
        self.salary+=5000
emp1=Employee("Sneha",50000)
emp1.increase_salary()
print(emp1.salary)
```

```python
class Book:
    def __init__(self,title,author):
        self.title=title
        self.author=author
    def details(self):
        print(f"Title:{self.title}")
        print(f"Author:{self.author}")
m1=Book("If you  die, you die","unknown")
m1.details()
```

```python
class Student:
    def __init__(self,name,marks):
        self.name=name
        self.marks=marks
    def average(self):
        return sum(self.marks)/len(self.marks)
s1=Student("Jimin",[90,85,95])
print(s1.average())
```
